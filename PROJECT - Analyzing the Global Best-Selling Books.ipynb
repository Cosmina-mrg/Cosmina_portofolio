```python
# 4. Clean the data
```

```python
# Remove dulplicates
df.drop_duplicates(inplace = True)
```

```python
df
```

```python
# Check if are null values in a Dataframe
df.isnull().values.any()
```

```python
# Rename the columns to make them more descriprive
df.rename(columns = {'Name': 'Title', 'Year': 'Publication Year', 'User Rating': 'Rating'}, inplace = True)
```

```python
df.head()
```

```python
# Convert data types
df['Price'] = df['Price'].astype(float)
```

```python
df.head()
```

```python
# 5. Exploratory Data Analysis
```

```python
# Which author have the most books on the best sellers list?
author_popularity = df['Author'].value_counts().reset_index()
author_popularity
```

```python
# Which author have the highest sales?
top_authors = df.groupby('Author')['Price'].sum().nlargest(10).reset_index()
top_authors
```

```python
# Average rating of every best selling author
average_rated_books = df[['Author', 'Title', 'Rating']].groupby(['Author', 'Title']).mean().reset_index()
average_rated_books
```

```python
# Top 10 rated books
top_rated_books = df[['Author', 'Title', 'Rating']].groupby(['Author', 'Title']).mean().sort_values(by = ['Rating'], ascending = False).head(10).reset_index()
top_rated_books
```

```python
# What is the lowest rating?
lowest_rating = df['Rating'].min()
lowest_rating
```

```python
# What is the title with the lowest rating?
lowest_rating_df = df[df['Rating'] == lowest_rating]
lowest_rating_df
```

```python
# Average Rating by Genre
average_rating_genre = df.groupby('Genre')['Rating'].mean()
average_rating_genre
```

```python
# The number of books by genre over the years
count_genre = df.groupby(['Publication Year','Genre']).size().reset_index(name ='Count')
count_genre
```

```python
# What is the highest and the lowest price?
max_price = df['Price'].max()
min_price = df['Price'].min()
print(max_price)
print(min_price)
```

```python
# How many books are sold at the highest price?
count_max = df[df['Price'] == 105]
print(count_max.size)
```

```python
# How many books are sold at the lowest price?
count_min = df[df['Price'] == 0]
print(count_min.size)
# There are 84 books for which price is 0. It may be because of those books being available in electronic format.
```

```python
# Display the information of the book that has the most reviews
maximum_reviews = df['Reviews'].max()
maximum_reviews_df = df[df['Reviews']== maximum_reviews]
maximum_reviews_df
```

```python
# 6. Data Visualization
```

```python
# Best selling genre
selling_genre = df['Genre'].value_counts()
ax = sns.barplot(x=selling_genre.values, y=selling_genre.index)
ax.bar_label(ax.containers[0],label_type = 'center')
plt.title('Best selling genre')
plt.show()
```

```python
# Top 10 best-selling authors
top_authors = df.groupby('Author')['Price'].sum().nlargest(10).reset_index()
plt.figure(figsize=(10, 5))
plt.barh(top_authors['Author'], top_authors['Price'])
plt.xlabel = ('Author')
plt.ylabel = ('Total sales')
plt.title = ('Top 10 best-selling authors')
plt.show()
```

```python
# Number of books over the years according to each genre
count_genre = df.groupby(['Publication Year','Genre']).size().reset_index(name ='Count')
fiction = count_genre[count_genre['Genre'] == 'Fiction']
nonfiction = count_genre[count_genre['Genre'] == 'Non Fiction']
fic = fiction['Count'].tolist()
nonfic = nonfiction['Count'].tolist()
year = fiction['Publication Year'].tolist()
group_count = pd.DataFrame({'Fiction': fic,'Non Fiction': nonfic}, index = year)
group_count.plot.bar()
```

```python
# Create a histogram for the Rating variables
sns.histplot(data = df, x = 'Rating', bins = 10, kde = True)
```
