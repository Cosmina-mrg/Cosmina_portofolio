```python
 # 1. Import the libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 2. Load the dataset

df = pd.read_csv(r'C:\Users\cosmi\OneDrive\Desktop\Python\Electric Vehicle Analysis\Electric Vehicle Population Data.csv')
```

```python
# 3. Inspect the dataset
df.info()
```

```python
print(df.shape)
```

```python
# Summary description
df.describe(include = 'all')
```

```python
# 4. Data Cleaning and Preprocesing
```

```python
# Remove some columns that are not important for study 
df.drop(['VIN (1-10)', 'Postal Code', 'Legislative District', 'DOL Vehicle ID'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
# Check the missing values
df.isnull().sum()
```

```python
# Remove missing data points
df = df.dropna()
df
```

```python
df.isna().sum()
```

```python
# Creating a new column 'Location'
df['Location'] = df['County']+ ',' + df['City'] + ',' + df['State']
df.head()
```

```python
# 5. Exploratory Data Analysis
```

```python
# How many electric vehicles are produced in each location?

location_count = df.groupby('Location').size().nlargest(10).reset_index(name = 'Electric Vehicle Count')
location_count.sort_values(by ='Electric Vehicle Count', ascending = False)
```

```python
location_count.set_index('Location', inplace = True)
plt.figure(figsize = (10,5))
location_count.plot.bar(title = 'Location Distribution',  color = 'DarkBlue')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Percentage of each location
location_count = df.groupby('Location').size().reset_index(name = 'Electric Vehicle Count')
total_vehicles = location_count['Electric Vehicle Count'].sum()
location_count['Percentage'] = (location_count['Electric Vehicle Count']/total_vehicles) * 100
location_count.sort_values(by ='Electric Vehicle Count', ascending = False)
```

```python
# Percentage of top 10 locations 

location_count.set_index('Location', inplace = True)
location_count['Electric Vehicle Count'].plot.pie(title = 'Percentage of location', autopct = '%1.1f%%', figsize = (10,8))
```

```python
# How many electric vehicles are produced per year?

counts_per_year = df['Model Year'].value_counts().sort_index().reset_index()
counts_per_year
```

```python
counts_per_year.set_index('Model Year', inplace = True)
plt.figure(figsize = (10,5))
counts_per_year.plot.bar(title ='Trend in Electric Vehicle Manufacturing Over Time', color = 'darkred')
plt.show()
```

```python
# What are the most electric vehicles manufacturer?

top_make = df['Make'].value_counts().head(10).reset_index(name = 'Count')
top_make.style.background_gradient()
```

```python
top_make.set_index('Make', inplace = True)
plt.figure(figsize=(10,5))
top_make['Count'].plot.barh(title = 'Top 10 electric vehicles manufacturer', color = 'darkgreen')
plt.gca().invert_yaxis()
plt.show()
```

```python
# What are the most electric vehicles models produced?

top_model = df['Model'].value_counts().head(10).reset_index(name = 'Count')
top_model.style.background_gradient()
```

```python
top_model.set_index('Model', inplace = True)
plt.figure(figsize = (10,5))
top_model.plot.barh(title = 'Top 10 models of the electric vehicles')
plt.gca().invert_yaxis()
plt.show()
```

```python
#Top 10 Most Popular Electric Vehicle Models by Manufacturer

models_make = df.groupby('Make')['Model'].value_counts().sort_values(ascending = False).head(10).reset_index(name = 'Number of vehicles')
models_make
```

```python
models_make.set_index('Make', inplace = True)
sns.barplot(data = models_make, x = 'Number of vehicles', y = 'Model', hue = 'Make')
plt.title('Top 10 Most Popular Electric Vehicle Models by Manufacturer')
plt.xlabel('Number of Vehicles')
plt.ylabel('Model')
plt.show()
```

```python
# How many vehicles are produced of each type?

type = df['Electric Vehicle Type'].value_counts().reset_index()
type
```

```python
type.set_index('Electric Vehicle Type', inplace = True)
myexplode = [0.2, 0]
type['count'].plot.pie(title = 'Most produced electric vehicle type', explode = myexplode, autopct = '%1.1f%%', shadow = True)
plt.show()
```

```python
# What are the rates of Eligibility(CAFV)?

count_of_eligibilty = df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].value_counts().reset_index()
count_of_eligibilty
```

```python
count_of_eligibilty.set_index('Clean Alternative Fuel Vehicle (CAFV) Eligibility', inplace = True)
plt.figure(figsize = (10,5))
count_of_eligibilty.plot.bar(title ='Rates of eligibility', color = 'skyblue')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# What are the most Electric Utility Companies?

electric_utility = df.groupby('Electric Utility').size().nlargest(10).reset_index(name = 'Count')
electric_utility.style.background_gradient()
```

```python
electric_utility.set_index('Electric Utility', inplace= True)
plt.figure(figsize= (10,5))
electric_utility.plot.bar(title = 'Top 10 Electric Utility Companies', color = 'burlywood')
plt.xticks(rotation = 45, ha= 'right')
plt.show()
```
